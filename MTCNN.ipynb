```python
import cv2
from mtcnn import MTCNN
```

```python
detector = MTCNN()
```

```python
image = cv2.imread("1cat.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
result = detector.detect_faces(image_rgb)
```

```python
for face in result:
    x, y, w, h = face['box']
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)


```
